```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data_v1 = pd.read_csv(r"C:\Users\Utilisateur\Desktop\Projet_interpromo_2k22\ProjectInterPromo\data\Donnees_v1.csv")
```

```python
variables = ['NBJ_CNT_SMS_6M',
 'TOT_CMS_WEB_6M',
 'NB_RETRAIT_MN_6M',
 'nb_op_cred_ep_a_terme_M',
 'ENCOURS_HABITAT',
 'nb_produit_Credit',
 'nb_DAV',
 'TOT_CHABITAT_WEB_6M',
 'NBJ_CNT_AGENCE_VENTE_6M',
 'TOT_VIR_WEB_6M',
 'nb_operation_debit_ep_a_vue_M',
 'NB_JR_MACARTE_6M',
 'nb_operation_credit_ep_a_vue_M6',
 'mt_paiement_chq_M',
 'NBJ_CNT_RDV_AGENCE_6M',
 'nb_produit_Services',
 'NBJ_CNT_AUTOMATES_DEPOT_6M',
 'NBJ_CNT_AGENCE_RETRAIT_6M',
 'nb_ferm_prod_6M']
```

```python
data = data_v1[variables]
```

```python
def analyse(df):
    for i in df.columns:
        print(f'la variance de la variable {str(i)} est {np.var(df[str(i)])}')
```

```python
analyse(data)
data
```

```python
def affichage_histogramme(df):
    
    for i in df.columns:
        plt.hist(df[str(i)])
        plt.title(f'histogramme de la variable {str(i)}')
        plt.show()
```

```python
affichage_histogramme(data)
```

```python
normalize = lambda x: (x - np.mean(x)) / np.sqrt(np.var(x))

data_normalize = data
for i in data.columns :
    data_normalize[str(i)] = normalize(data[str(i)])
```

```python
analyse(data_normalize)
```
