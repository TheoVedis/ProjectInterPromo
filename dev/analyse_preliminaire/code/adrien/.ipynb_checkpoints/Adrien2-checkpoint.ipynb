# Fct bool si variable négative

```python
"""Entrée : Une colonne d'un dataframe
Sortie : Boolean True / False

premiere focntion qui cherche toutes les variables avec valeurs négatives (gerer cas par cas) 
=> True : valeure négative présentes
=> False : valeure positive"""
```

```python
def var_neg(colonne):
    for elem in colonne:
        if elem < 0:
            return True
    return False
```
