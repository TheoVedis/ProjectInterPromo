## Traitement des données quantitatives

```python
dfQuanti = df.drop(QualiColumns, axis=1, errors='ignore')
```

```python
for col in dfQuanti:
    dfTemp = dfQuanti[col]
    temp = dfTemp.isna().sum()/dfTemp.shape[0]
    if (temp > 0.10):
        print(dfTemp[:0].name + " : " + str(temp))
    
```

```python
dfQuanti = dfQuanti.fillna(value=0)
```

## Traitement des colonnes à donées qualitatives

```python
def get_seconds(time_str):
    # split in hh, mm, ss
    hh, mm, ss = time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)
```

```python
#Changing hour data to seconds
dfQuali['DUREE_CNX_WEB_TOT_6M'] = dfQuali['DUREE_CNX_WEB_TOT_6M'].fillna(value='00:00:00')
i=0
for val in dfQuali['DUREE_CNX_WEB_TOT_6M']:
    val_sec = get_seconds(val)
    dfQuali.loc[i,'DUREE_CNX_WEB_TOT_6M'] = int(val_sec/60)
    i+=1
            
```

```python
dfQuali['DUREE_CNX_MOY_WEB_6M'] = dfQuali['DUREE_CNX_MOY_WEB_6M'].fillna(value='00:00:00')
i=0
for val in dfQuali['DUREE_CNX_MOY_WEB_6M']:
    val_sec = get_seconds(val)
    dfQuali.loc[i,'DUREE_CNX_MOY_WEB_6M'] = int(val_sec/60)
    i+=1
```

```python

dfQuali = pd.get_dummies(dfQuali, columns=['seg_nature_potentiel', 'type_residence_1', 'situ_fam_1', 'contrat_travail_1', 'Tranche_age_1', 'sexe_1', 'Tranche_age_moyen', 'situ_globale', 'SGMT_FIDELITE', 'Fidelite_calcule', 'profil', 'localisation', 'INDICMAILING', 'INDICSMS', 'INDICPOR', 'INDICTEL', 'INDICMAIL', 'SECTEUR_GESTION'])
```

```python
dfQuali = dfQuali.drop(['cp_ou_pays'], axis=1)
```

```python
dfClear = dfQuanti.merge(dfQuali, how='inner', left_on='ID', right_on='ID')
```
